<h1>Seasonal Analysis</h1>

```python
# Accidents per year
accidents_per_year = roadacc['Year'].value_counts().sort_index()
accidents_per_month = roadacc['Month'].value_counts().sort_index()
accidents_per_day = roadacc['DayOfWeek'].value_counts().sort_index()


#PER YEAR 
accidents_per_year.plot(kind='bar')
plt.title('Accidents per Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.show()

month_names = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}

# Replace numerical months with their names
accidents_per_month.index = accidents_per_month.index.map(month_names)


accidents_per_month.plot(kind='bar')
plt.title('Accidents per Month')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.show()


#accident per day
day_names = {
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday'
}

accidents_per_day.index = accidents_per_day.index.map(day_names)
accidents_per_day.plot(kind='bar')
print(accidents_per_day)
```

<h3>1. Hillingdon Analysis</h3>
<p>Monthly Road Accidents analysis by year 2019-2022</p>

<h1>Using District Area to make a comparison to a general result</h1>

```python
Hillingdon_df = roadacc[roadacc['District Area'] == 'Hillingdon']

monthly_Hillingdondata = Hillingdon_df.groupby(['Year', 'Month']).size()

monthly_Hillingdondata = monthly_Hillingdondata.unstack(level=0)

monthly_Hillingdondata.plot(
    marker='o', 
    figsize=(10, 5),
    title='Monthly Road Accidents in Hillingdon by Year (2019-2022)'
)
plt.xticks(range(1, 13), ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.legend(title='Yearly Accident Comparison', bbox_to_anchor = (1.05,1), loc='upper left')
plt.grid(alpha=0.3)
plt.show()
```

<h3>Insights 1: On the year 2022 the month of september has the highest rate of road accidents ranging from 90.</h3>
<h3>Insigths 2: The year 2021 has the second highest rate of monthly road accidents on the month of june. On the same time it has the lowest rate of accidents on the month of August ranging to 40.</h3>

<h1>Accidents per year/month/day</h1>

```python
# Accidents per year
accidents_per_year = Hillingdon_df['Year'].value_counts().sort_index()
accidents_per_month = Hillingdon_df['Month'].value_counts().sort_index()
accidents_per_day = Hillingdon_df['DayOfWeek'].value_counts().sort_index()


#PER YEAR 
accidents_per_year.plot(kind='bar')
plt.title('Accidents per Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.show()

month_names = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'September', 10: 'October', 11: 'November', 12: 'December'
}

# Replace numerical months with their names
accidents_per_month.index = accidents_per_month.index.map(month_names)


accidents_per_month.plot(kind='bar')
plt.title('Accidents per Month')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
plt.show()


#accident per day
day_names = {
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday'
}

accidents_per_day.index = accidents_per_day.index.map(day_names)
accidents_per_day.plot(kind='bar')
print(accidents_per_day)
```

<h1>Accident Severity in Hillingdon</h1>

```python
severity_Hillingdon  = Hillingdon_df['Accident_Severity'].value_counts()

plt.pie(severity_Hillingdon,
        labels = severity_Hillingdon.index,
        autopct = '%1.1f%%',
        startangle = 90)
plt.gcf().set_size_inches(10,7)
plt.title('Accident Severity Distribution')
plt.show()

severity_2019_Hillingdon = Hillingdon_df[Hillingdon_df['Year'] == 2019]
# severity_2019
lsc_2019 = severity_2019_Hillingdon['Accident_Severity'].value_counts()

plt.pie(lsc_2019,
        labels = lsc_2019.index,
        autopct = '%1.1f%%',
        startangle = 90)
plt.gcf().set_size_inches(10,7)
plt.title('Accident Severity Distribution')
plt.show()
```
